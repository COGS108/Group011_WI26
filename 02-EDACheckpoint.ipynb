# Research Question

How do the seasons of the year impact the type of music that appears on Billboard top 100 songs weekly in terms of valence, tempo, and energy, and how do these features compare to one another on a seasonal basis?


## Background and Prior Work

Music is grounded in *a priori* structures of human experience, so it goes. In essence, music is one of the most universal and accessible ways humans, and societies as a whole, encounter emotion and being. As Tomas Clifton describes in their work *Music and the a Priori*, musical experience contains its own form of understanding that is **immediately given** while listening<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). These structuures are given in forms such as temporal flow, tension, and expectation, which are conditions of our existing in the world; They are not learned. It is because of this that we can call music a type of emotional landcape- A landscape that reveals how we experience and inhabit time, how we move through feeling, and how our inner states are susceptible to shift moment to moment, song-to-song. 

This persnal and deep connection between human emotion and music could offer an explanation for why some songs are described as energetic, uplifting, depressing, or calming<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). This account ***humanizes*** these labels, reflecting on how the music resonates with our own experiental structures, rather than seeming correlational and abstract. For example, faster tempos tend to evoke feelings of being upbeat or energetic while slower tempos can feel more calming or melancholic. Additionally, valence could lead us to label a song as either sad or happy, seemingly from our own intuition. Clifton's paper helps explain this intuitiveness, where music is directly tied to emotional meaning through structures that shape our experience of the world. 

The major point that our research project aims to address is the fact that emotional experience does not stay constant throughout the year. In the past couple decades, we have better come to understand environmental psychology by studying how seasonal changes in daylight, temperature, and daily routines can influence mood and motivation. The most obvious example of this comes from the term **winter depression**, a.k.a. Seasonal Affective Disorder. 

As Yale School of Medicine describes, SAD is characterized by dperessed mood, decreased enjoyment of activities, sadness, poor energy, and decreased confidence. Most importantly for our research question, through questionaires they note that roughly 90% of people claim to report changes in energy, modd, sleep, and apetite in the winter<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). This pattern typically starts in fall, seemingly creating a pattern of a rise and decline of mood and emotion that mimicks the dynamics of music itself with the cycles of intensity, valence, and tone that rise and fall over time. 

**Pettijohn et al. (2010) Music for the Seasons: Seasonal Music Preferences in College Students** is a central piece of prior work that directly links seasons to musical preference. In their paper, they describe two studies with undergraduate participants, where students were primed to have different seasons in mind. In the first study, they used a winter vs. summer reading scenario, and in the second they had students write about their own fall, winter, spring, or summer experiences. They then reported their preffered types of music, where the results supported the researchers' hypotheses that seasonal context would shift preferences toward reflexive and complex music in fall and winter, and toward energetic and upbeat music for spring and summer<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). This paper provides a strong foundation for investigating whether large-scale music trends from the Billboard Top100 also follow this shift from across different seasons. 

There is, however, some opposition to this theory. **Winthorst et al. (2020) Seasonality of mood and affect in a large general population sample** is a clear counter example to our premise- that seasons influence mood. The researchers studied mood and emotional well-being in adults to see if there were real differences in mood across the seasons. They measured positive and negative affect and depressive symptoms using questionaires. They found that season-to-season differences were very small and the evidence was not strong enough to say that seasons dramatically change mood for the great majority of people. The only meaningful effect was due to the personality trait neuroticism, but with only modest effects<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5). This, very clearly, seems to contradict our earlier claim on which we based our research project. While we cannot directly test to confirm or reject their findings, if our results yield a difference of audio features that simmilarly match the findings of Pettijohn et al. (2010), then we may be able to assume some relationship between the two.

### Dataset #1: Billboard Hot 100 (2020–2025 Subset)
Dataset Name: Billboard Hot 100 Complete Chart History
Link to Dataset: https://github.com/mhollingshead/billboard-hot-100
Number of Variables: 7

This dataset originally contains weekly Billboard Hot 100 rankings from 1958 to the present. For this project, the dataset was filtered to include only data from 2020–2025 in order to focus on modern music trends in the streaming era.

Each observation represents one song on the Billboard Hot 100 chart for a specific week.

The most relevant variables include:

song – The name of the song.
artist – The performing artist.
date – The chart week date.
this_week – The song’s rank that week (1–100).
last_week – The song’s previous week rank.
peak_position – The highest rank the song has reached.
weeks_on_chart – Total weeks the song has appeared on the chart.

Chart position is measured from 1 to 100, where 1 indicates the most popular song that week.

One limitation of this dataset is that it does not include musical audio features such as tempo (beats per minute), energy, loudness (decibels), danceability, or valence (musical positivity). Because of this, it may need to be combined with a dataset such as Spotify audio features to analyze musical characteristics. Additionally, differences in how artist names are formatted (for example, featured artists) may make merging datasets more challenging.

```python
## CODE TO LOAD THE DATA
import json
import pandas as pd

with open('data/00-raw/billboard_all.json') as f:
    data = json.load(f)

rows = []

for chart in data:
    chart_date = chart["date"]
    for song in chart["data"]:
        song["date"] = chart_date
        rows.append(song)

billboard_df = pd.DataFrame(rows)
```

```python
billboard_df["date"] = pd.to_datetime(billboard_df["date"])

##Filter for years 2020–2025
billboard_df = billboard_df[
    (billboard_df["date"].dt.year >= 2020) &
    (billboard_df["date"].dt.year <= 2025)
]

billboard_df.shape
```

```python
billboard_df.shape
billboard_df.head()
```

```python
billboard_df.isnull().sum()
```

```python
def get_season(date):
    month = date.month
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Fall"

billboard_df["season"] = billboard_df["date"].apply(get_season)

billboard_df.tail()
```

### Dataset #2: Spotify Song Features to Billboard Hot 100
**Dataset Description**

The Spotify audio features dataset provides a matching set of quantitative measurements that describe the musical and acoustic properties of unique tracks. These features are produced by Spotify's internal data and algorithmic analysis to extract standardized metrics. For this project, our primary focus in three variables- valence, energy, and tempo- which captures different dimensions of musical expression. 

Valence and Energy are unitless scores ranging from 0.0 to 1.0, where valence represents the "positivity" of a track and energy measures intesnity and activity level of the track. Songs with high valence will sound happy and/or cheerful, while high-energy songs will feel loud, fast, and dynamic. Tempo is measured in units of beats oer minute (BPM) and reflects the speed of the underlying beat. Together, these metrics will allow us to quanitify emotional and structural aspects of music across different seasons. 

Although valence, energy, and tempo are the core features for this project, Spotify offers a wide range of additional metrics like danceability, loudness, mode, speechiness, acousticness, etc. These features are not used directly in the project analysis, but are part of the audio features that will be extracted.

This dataset was constructed by merging two publicly available Kaggle datasets with the Billboard Hot 100 dataset shown in Dataset #1. These datasets, *Billboard Hot 100 Audio Features* and *Billboard 24 Years Lyrics & Spotify Features*, only cover weekly chart data from 2020 to 2023. The merge was performed on normalized song titles and primary aritst name only to improve match rates. The final matched dataset contained roughly 6800 rows, representing Billboard chart appearances with Spotify audio feautres, out of 30,000+ rows across the original 2020-2025 dataset. 


**Dataset Concerns and Limitations**

There are key limitations that must be considered when working with this type of data. First is the software aspect through API rate limits. Spotify restricts the number of requests per second which can delay or block data collection, which can delay or block data collection for days at a time. More importantly to our project, Spotify deprecated its audio features API endpoint, preventing mass retrieval of features for songs charting in 2024-2025, forcing our reliance on pre-existing Kaggle datasets instead. Our second concern is the low match rate of only 21.5%. Additionally, we have proxy limitations of this data, where valence and energy are algorithmic estimates, not direct measurements of human emotions.

```python
import pandas as pd
import re

# Load both Kaggle datasets
hot100_df   = pd.read_csv("Hot_100_Audio_Features.csv")
lyrics_df   = pd.read_csv("billboard_24years_lyrics_spotify.csv")

hot100_df.tail(3)
```

#### Normalizing the Dataframes
You might have noticed the difference in how the hot100_df and the billboard_df treat secondary and featuring artists. We must ensure that the names are one-to-one with eachother across datasets to make sure they merge properly since we pair each song with its features using both song and artist names. 

```python
# strip all secondary artist credits
def strip_featuring(s):
    if not isinstance(s, str):
        return s
    # Strip feat/ft/featuring first
    s = re.split(r'\s+(feat\.?|ft\.?|featuring)\s+', s, flags=re.IGNORECASE)[0]
    # Strip secondary artists after &, x, with, +
    s = re.split(r'\s+(&|x|with|\+)\s+', s, flags=re.IGNORECASE)[0]
    return s.strip()
hot100_df["song_clean"]   = hot100_df["Song"].str.lower().str.strip().apply(strip_featuring)
hot100_df["artist_clean"] = hot100_df["Performer"].str.lower().str.strip().apply(strip_featuring)

lyrics_df["song_clean"]   = lyrics_df["song"].str.lower().str.strip().apply(strip_featuring)
lyrics_df["artist_clean"] = lyrics_df["band_singer"].str.lower().str.strip().apply(strip_featuring)

# Build combined lookup table (song + artist)
hot100_lookup = hot100_df[["song_clean", "artist_clean", "valence", "energy", "tempo"]] \
    .dropna(subset=["valence", "energy", "tempo"])

lyrics_lookup = lyrics_df[["song_clean", "artist_clean", "valence", "energy", "tempo"]] \
    .dropna(subset=["valence", "energy", "tempo"])

combined_lookup = pd.concat([hot100_lookup, lyrics_lookup], ignore_index=True) \
    .drop_duplicates(subset=["song_clean", "artist_clean"], keep="first")

print(f"Combined lookup table: {len(combined_lookup)} unique songs")

# Normalize full billboard_df
billboard_df["song_clean"]   = billboard_df["song"].str.lower().str.strip().apply(strip_featuring)
billboard_df["artist_clean"] = billboard_df["artist"].str.lower().str.strip().apply(strip_featuring)
```

#### Merge and Final Dataframe
Now that the names are normalized across dataframes, it is now time to merge the Kaggle datasets into our main Billboard dataframe. 

```python
# Merge on song + artist
final_df = billboard_df.merge(
    combined_lookup,
    on=["song_clean", "artist_clean"],
    how="left"
)

matched_p1 = final_df["valence"].notna().sum()
print(f"Matched (song + artist): {matched_p1}/{len(final_df)} rows ({100*matched_p1/len(final_df):.1f}%)")

final_df = final_df.drop(columns=["song_clean", "artist_clean"])

# Sanity check
assert len(final_df) == len(billboard_df), (
    f"Row count mismatch! Expected {len(billboard_df)}, got {len(final_df)}."
)
print(f"Row count matches: {len(final_df)} rows")

# match rate
matched = final_df["valence"].notna().sum()
total   = len(final_df)
print(f"Total matched: {matched}/{total} rows ({100*matched/total:.1f}%)")

# Drop unmatched rows
final_df = final_df.dropna(subset=["valence", "energy", "tempo"])
print(f"Remaining after drop: {final_df.shape}")

final_df.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import warnings
import numpy as np
warnings.filterwarnings('ignore')

#Load from processed file
final_df = pd.read_csv("data/02-processed/final_merged.csv")
df = final_df.copy()
df["date"]  = pd.to_datetime(df["date"])
df["month"] = df["date"].dt.month
df["year"]  = df["date"].dt.year

#Shared style constants
SEASON_ORDER  = ["Winter", "Spring", "Summer", "Fall"]
SEASON_COLORS = {
    "Winter": "#5B8DB8",
    "Spring": "#6DBF67",
    "Summer": "#F4A621",
    "Fall":   "#C1603A"
}
palette = [SEASON_COLORS[s] for s in SEASON_ORDER]
sns.set_theme(style="whitegrid", font_scale=1.1)
plt.rcParams.update({"figure.dpi": 130, "axes.spines.top": False, "axes.spines.right": False})

print(f"Working dataset: {len(df):,} matched chart entries (2020–2023)")
print(f"\nSeason breakdown:\n{df['season'].value_counts().reindex(SEASON_ORDER)}")
print(f"\nSummary statistics:")
print(df[["valence", "energy", "tempo"]].describe().round(3))
```

```python
#CELL 1 - SECTION 1: Chart entries per season
#Before comparing features across seasons we verify the matched data
#is reasonably balanced. An uneven distribution could bias our seasonal
#comparisons given the 21.5% match rate from merging Billboard with the Kaggle
#feature datasets.

counts = df['season'].value_counts().reindex(SEASON_ORDER)

fig, ax = plt.subplots(figsize=(7, 5))
bars = ax.bar(SEASON_ORDER, counts.values, color=palette, alpha=0.85, edgecolor="white", linewidth=1.5)
for bar, val in zip(bars, counts.values):
    ax.text(bar.get_x() + bar.get_width()/2, val + 10,
            f"{val:,}", ha="center", va="bottom", fontsize=10)

ax.set_ylabel("Number of Matched Chart Entries")
ax.set_title("Matched Billboard Chart Entries per Season\n(2020–2023)", fontweight="bold")
plt.tight_layout()
plt.savefig("data/02-processed/plot_counts.png", bbox_inches="tight")
plt.show()
```

```python
# CELL 2 - SECTION 1: Violin plots
# Violin plots show the full distribution shape of each audio
# feature per season, not just the mean. This matters because
# valence and energy tend to be skewed in real music data.
# The inner box shows the median and interquartile range.
# We look for whether Winter distributions skew lower for
# valence and energy compared to Summer.

fig, axes = plt.subplots(1, 3, figsize=(16, 6))

for ax, feat, ylabel, title in zip(
    axes,
    ["valence", "energy", "tempo"],
    ["Valence (0–1)", "Energy (0–1)", "Tempo (BPM)"],
    ["Valence by Season", "Energy by Season", "Tempo by Season"]
):
    sns.violinplot(
        data=df, x="season", y=feat,
        order=SEASON_ORDER, palette=palette,
        inner="box", linewidth=1.2, ax=ax
    )
    ax.set_title(title, fontweight="bold", pad=10)
    ax.set_xlabel("")
    ax.set_ylabel(ylabel)

fig.suptitle(
    "Distribution of Audio Features Across Seasons\n(Billboard Hot 100, 2020–2023)",
    fontsize=14, fontweight="bold", y=1.01
)
plt.tight_layout()
plt.savefig("data/02-processed/plot_violins.png", bbox_inches="tight")
plt.show()
```

```python
# CELL 3 - SECTION 1: Grouped bar chart (medians)
# A direct summary of median valence, energy, and tempo side
# by side across all four seasons. We use median rather than
# mean because the violin plots revealed skewed distributions.
# Tempo is normalized to 0-1 for visual comparison alongside
# valence and energy on the same scale.

season_medians = df.groupby("season")[["valence", "energy"]].median().reindex(SEASON_ORDER)
tempo_medians  = df.groupby("season")["tempo"].median().reindex(SEASON_ORDER)
tempo_norm     = (tempo_medians - tempo_medians.min()) / (tempo_medians.max() - tempo_medians.min())
season_medians["tempo_norm"] = tempo_norm.values

x     = np.arange(len(SEASON_ORDER))
width = 0.25

fig, ax = plt.subplots(figsize=(10, 6))
b1 = ax.bar(x - width, season_medians["valence"],    width, label="Valence",            color="#E84393", alpha=0.85)
b2 = ax.bar(x,          season_medians["energy"],     width, label="Energy",             color="#F4A621", alpha=0.85)
b3 = ax.bar(x + width,  season_medians["tempo_norm"], width, label="Tempo (normalized)", color="#5B8DB8", alpha=0.85)

ax.set_xticks(x)
ax.set_xticklabels(SEASON_ORDER, fontsize=12)
ax.set_ylabel("Median Score (0–1)")
ax.set_title("Median Audio Features by Season\n(Billboard Hot 100, 2020–2023)", fontweight="bold")
ax.legend()
ax.set_ylim(0, 1.0)

for bars in [b1, b2, b3]:
    for bar in bars:
        h = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, h + 0.01,
                f"{h:.2f}", ha="center", va="bottom", fontsize=8.5)

plt.tight_layout()
plt.savefig("data/02-processed/plot_grouped_bars.png", bbox_inches="tight")
plt.show()
```
